```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt
X_file = pd.read_csv("motor_temperature_x_test.csv")
y_file = pd.read_csv("motor_temperature_y_test.csv")
```

```python
loaded_rg = joblib.load("motor_temperature.joblib")

X = X_file.iloc[:,1:]
#X.head()

y = y_file.iloc[: ,1:]
#y.head()
```

```python
pred_y= loaded_rg.predict(X)

x = [i for i in range(500)]
plt.plot(x, pred_y[500:1000], color = 'red')
plt.plot(x, y[500:1000], color = 'green')
plt.show()
```
